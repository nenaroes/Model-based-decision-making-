```python
# Load in the data for the executed runs
from ema_workbench import load_results

experiments, results = load_results('8000 scenarios 50 policies.tar.gz')
```

```python
# Create dataframes of the experiments and the results
df_experiments = pd.DataFrame.from_dict(experiments)
df_results = pd.DataFrame.from_dict(results)
```

```python
# Keep the desired columns for the experiments dataframe
df_experiments = df_experiments[['A.0_ID flood wave shape',
 'A.1_Bmax',
 'A.1_Brate',
 'A.1_pfail',
 'A.2_Bmax',
 'A.2_Brate',
 'A.2_pfail',
 'A.3_Bmax',
 'A.3_Brate',
 'A.3_pfail',
 'A.4_Bmax',
 'A.4_Brate',
 'A.4_pfail',
 'A.5_Bmax',
 'A.5_Brate',
 'A.5_pfail',
 'discount rate',
 '0_RfR',
 '1_RfR',
 '2_RfR',
 '3_RfR',
 '4_RfR',
 'A.1_DikeIncrease',
 'A.2_DikeIncrease',
 'A.3_DikeIncrease',
 'A.4_DikeIncrease',
 'A.5_DikeIncrease',
 'EWS_DaysToThreat',
 'scenario_id',
 'policy']]
```

```python
# Create a total dataframe of the experiments and the results
# This dataframe is just created to get an idea of the results that belong to certain experiments
df_total = pd.concat(objs = [df_experiments, df_results], axis = 1)
df_total
```

```python
# Calculate the mean of the expected number of deaths and of all costs 
mean_deaths = df_total['Expected Number of Deaths'].mean()
print(mean_deaths)

mean_costs = df_total['All Costs'].mean()
print(mean_costs)
```

```python
# Calculate the maximum of the expected number of deaths and of all costs 
mean_deaths = df_total['Expected Number of Deaths'].max()
print(mean_deaths)

mean_costs = df_total['All Costs'].max()
print(mean_costs)
```

```python
# Calculate the minimum of the expected number of deaths and of all costs 
mean_deaths = df_total['Expected Number of Deaths'].min()
print(mean_deaths)

mean_costs = df_total['All Costs'].min()
print(mean_costs)
```

```python
# Create two different dataframes of the two outcomes of interest
df_results_costs = df_results[['All Costs']]
df_results_deaths = df_results[['Expected Number of Deaths']]
```

```python
# Create a boxplot of the outcomes for 'Expected Number of Deaths'
df_results_deaths.boxplot()
plt.show()
```

```python
# Create a boxplot of the outcomes for 'All Costs'
df_results_costs.boxplot()
plt.show()
```

## 17 objectives

```python
# Load in the data for the executed runs
from ema_workbench import load_results

experiments, outcomes = load_results('4000 scenarios 25 policies 17 objectives.tar.gz')
```

```python
# Preparing a dataframe that can be used to calculate the correlations between the outcomes of interest
data = {}
for k,v in outcomes.items(): 
    
    if len(v.shape) > 1:
        v = v[:,-1]
    data[k] = v

data = pd.DataFrame.from_dict(data)
```

```python
# Calculating the correlations between the outcomes of interest
data.corr()
```

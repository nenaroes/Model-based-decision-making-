```python
# Importing the standard Python packages
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading in the results from the experiments computed for the feature scoring
from ema_workbench import load_results

experiments, outcomes = load_results('3 obj 4000 scenarios 50 policies.tar.gz')
```

```python
# Executing the feature scoring analysis included in the EMA workbench analysis
from ema_workbench.analysis import feature_scoring
 
x = experiments
y = outcomes

fs = feature_scoring.get_feature_scores_all(x, y)

fig, ax = plt.subplots(figsize=(15,15))         # Sample figsize in inches
sns.heatmap(fs, cmap='viridis', annot=True)

plt.show()
```

```python
# Saving the figure presented above
fig.savefig('FeatureScoring3obj.png')
```
